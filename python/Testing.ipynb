```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt
import random
from scipy import stats
from growthrates import *
```

```python
N = 11
Steps = 3
sites = np.arange(N)
S = reset_S(N, 0)
plt.plot(sites, S)
for i in range(Steps):
    S = brick_update_S(S)
    plt.plot(sites, S)
# print(np.shape(heights))

plt.show()
```

```python
N = 10
Steps = 10
sites = np.arange(N)
S = reset_S(N, 0)
plt.plot(sites, S)
for i in range(Steps):
    S = random_update_S(S)
    plt.plot(sites, S+ i/10)
# print(np.shape(heights))

plt.show()
```

```python
N = 10
Steps = 20
sites = np.arange(N)
S = reset_S(N, 0)
plt.plot(sites, S)
for i in range(Steps):
    S = stair_update_S(S,3)
    plt.plot(sites, S+ i/20)
# print(np.shape(heights))

plt.show()
```

```python
N = 2500
n = 1200
Steps = 4
diff = 600
sites = np.arange(N+1)
S = np.arange(N+1)
S = S%2
plt.plot(sites, S)
for i in range(Steps):
    for j in range(diff):
        S = stair_update_S_fixed(S,n)
    plt.plot(sites, S)
# print(np.shape(heights))

plt.savefig('../figures/large_stair_growth.pdf')
plt.show()
```
