```python
import numpy as np
```

```python
two_max = np.array([[-2, 2, 1, 0, 0, 0],
                    [ 1,-4, 1, 1, 0, 0],
                    [ 1, 1,-4, 1, 1, 0],
                    [ 0, 1, 1,-4, 1, 1],
                    [ 0, 0, 1, 1,-3, 1],
                    [ 0, 0, 0, 1, 1,-2]])
np.linalg.eig(two_max)
```
