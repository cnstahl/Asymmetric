```python
import numpy as np
import asymmetric as asym
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
L = 9; end = 3; n = 20

weights = asym.get_all_weights(L, end, n, here=True, pauli=True, dense=True,
                               dot_strength=None, field_strength=1,
                               Azero=True, Aplus=True, Amult=False)
```

```python
print(weights.shape)
```

```python
for i in weights:
    for j in i:
        plt.plot(j)
    plt.show()
```

```python
stop = (int) (end*n)
fore, back = weights[4:6]

ax = plt.subplot(111)
for i in range(L):
    ax.plot(np.arange(stop)/n,fore[i,:stop], label = str(i))
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.9, box.height])
ax.legend(bbox_to_anchor=(1, 1), loc=2)
plt.ylim(0,1)
plt.ylabel('Forward OTOC')
plt.xlabel('site')
plt.title('Without Disorder')
plt.savefig('figures/plusOTOCnoDisorder_fore.pdf')
plt.show()

ax = plt.subplot(111)
for i in range(L):
    ax.plot(np.arange(stop)/n,back[i,:stop], label = str(i))
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.9, box.height])
ax.legend(bbox_to_anchor=(1, 1), loc=2)
plt.ylim(0,1)
plt.ylabel('Backward OTOC')
plt.xlabel('site')
plt.title('Without Disorder')
plt.savefig('figures/plusOTOCnoDisorder_back.pdf')
plt.show()
```
